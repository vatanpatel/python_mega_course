```python
import folium
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('volcanos.txt')
```

```python
lon = list(data['LON'])
lat = list(data['LAT'])
elev = list(data['ELEV'])
```

```python
def fn(x):
    if x <3000:
        return('green')
    elif x < 4000 and x >= 3000:
        return('pink')
    else:
        return('red')
```

```python
map = folium.Map(location = [np.mean(lat), np.mean(lon)], zoom_start=6)
        
```

```python
fg1 = folium.FeatureGroup(name = 'Volcanoes')
```

```python
for x, y, z in zip(lat, lon, elev):
    fg1.add_child(folium.CircleMarker(location= [x,y], radius = 3, popup = z, fill_color = fn(z), color = 'gray', fill_opacity = 0.7))
```

```python
fg2 = folium.FeatureGroup(name = 'Population')
```

```python
fg2.add_child(folium.GeoJson(data = open('world.json', 'r', encoding='utf-8-sig').read(), style_function = lambda x: {'fillColor':'green' if x['properties']['POP2005'] < 10000000 else 'orange' if 10000000 <= x['properties']['POP2005'] < 20000000 else 'red'}))
```

```python
map.add_child(fg1)
map.add_child(fg2)
map.add_child(folium.LayerControl())
```

```python
map.save("Map.html")
```
